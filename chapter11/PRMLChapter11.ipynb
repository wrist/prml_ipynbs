## この章での基本的な問題 ##

* 非観測変数の事後分布はほとんどの場合、予測を行うための期待値を評価される目的で必要とされる
* この章で触れる内容
    * ある関数$f({\bf z})$の確率分布$p({\bf z})$の下での期待値の計算に関すること
        * ${\bf z}$の要素は離散、連続、それらの組み合わせもある
    * 連続変数の場合は下記の式(離散の場合は和になる)
        * この期待値は解析的に厳密に評価することは不可能であると仮定

$$ \mathbb{E}[f]=\int{f({\bf z})p({\bf z})d{\bf z}} $$


## サンプリング法のアイデア ##

* 分布$p({\bf z})$から独立に抽出されたサンプルの集合${\bf z}^{(l)} (l=1,...,L)$を得ること
* これにより期待値を有限和で近似可能
$$ \hat{f}=\frac{1}{L}\sum_{l=1}^L f({\bf z}^{(l)}) $$

* サンプル${\bf z}^{(l)}$が分布$p({\bf z})$に従って抽出される限り、$\mathbb{E}[\hat{f}]=\mathbb{E}[f]$が成り立ち、推定量$\hat{f}$は正しい平均を持つ

## 推定量の分散 ##

$$ {\rm var}[\hat{f}]=\frac{1}{L}\mathbb{E}[(f-\mathbb{E}[f])^2] $$

これは関数$f({\bf z})$の分布$p({\bf z})$の下での分散である(演習11.1)

## 推定量の精度 ##

* ${\bf z}$の次元数に依存しない
* 原理的には比較的少数のサンプル${\bf z}^{(l)}$で高い精度が達成可能
    * 実際には10〜20の独立なサンプルで十分

## サンプリングにおける問題 ##

* サンプル${\bf z}^{(l)}$が独立でないかもしれない
    * 実効的なサンプルサイズが見かけのサンプルサイズよりもずっと少ないかもしれない
* 図11.1の例
    * $p({\bf z})$が大きい領域で$f({\bf z})$が小さく、逆に$p({\bf z})$が小さい領域で$f({\bf z})$が大きい場合
         * 期待値は小さな確率の領域から大きな影響を受ける
    * 十分な精度を得るためには比較的大きなサンプルサイズが必要

### アルゴリズム ###

1. ${\bf z}_i$のサンプル値と観測値と比較

### ロジックサンプリングの特徴 ###

* 事後分布から正しくサンプリングを行う
    * 隠れ変数とデータ変数の同時分布からサンプリングし、観測データと一致しないサンプルを破棄することと同じであるため
    * 矛盾する値の観測時はすぐに同時分布からのサンプリングを打ち切るためわずかに時間を節約
* 観測変数の数と取りうる状態数が増えるに従って事後分布からのサンプルを受理する全体的な確率は急速に減少
    * そのためこのアプローチが使われることはめったにない

# 11.1 基本的なサンプリングアルゴリズム #

## 11.1.1 標準的な分布 ##

* zが区間(0,1)で一様に分布し、ある関数f(・)を用いてzの値をy=f(z)のように変換する
* yの分布
$$ p(y) = p(z) \left|\frac{{\rm d}z}{{\rm d}y} \right| $$

* ここで一様分布であるためp(z)=1

## 変換後のyの値が求めたい特定の分布p(y)に従うようにf(z)を選ぶ ##

* 先の式を積分することで以下を得る

$$ z = h(y) \equiv \int_{-\infty}^y p(\hat{y})d\hat{y}$$

* $y=h^{-1}(z)$
    * 求めたい分布の不定積分の逆関数を用いて一様分布の乱数を変換する必要がある

## 演習11.2 ##

(11.6)式
$$ z = h(y) \equiv \int_{-\infty}^y p(\hat{y})d\hat{y}$$
および$y=h^{-1}(z)$を前提とした時にyが分布p(y)を持つことを示せ

### 答え ###
(11.6)式をyで微分すると
$$ dz/dy = p(y) $$
ここでy=f(z)と変換したときのyの確率分布は下記であるが
$$ p(y) = p(z) \left|\frac{{\rm d}z}{{\rm d}y} \right| $$
一様分布の場合p(z)=1であるのでdz/dyを代入すると両辺一致し、yはp(y)を分布に持つことが分かる。

## 指数分布(exponential distribution)の例 ##

$$ p(y) = \lambda \exp (-\lambda y) $$
ここで$0\leq y < \infty$である。

* 指数分布の場合不定積分の下界は0であるため不定積分を計算すると

$$ \int_0^y \lambda \exp (-\lambda y) dy = \left[ (-\lambda^{-1}) \lambda \exp (-\lambda y) \right]_0^y = 1 - \exp(-\lambda y)$$

* この逆関数を求めればそれを用いて指数分布に従うサンプルを得ることができる

$$ z = h(y) = 1 - \exp(-\lambda y)$$
$$ \iff \exp(-\lambda y) = 1 - z $$
$$ \iff -\lambda y = \ln (1 - z) $$
$$ \iff y = -\lambda^{-1} \ln (1-z) $$

```python
def exp_dist(xs, exp_param=1.0):
    return exp_param * np.exp(- exp_param * xs)

def inv_integral_of_exp_dist(zs, exp_param=1.0):
    return - np.log(1.0 - zs) / exp_param
```

```python
from numpy.random import RandomState

# 指数分布
xs = np.linspace(0.0, 10.0, 1000)
exps = exp_dist(xs)

# 0.0 ~ 1.0の一様分布に従うサンプルを1000点生成
# RandomStateの引数はseed
zs = RandomState(1).uniform(0.0, 1.0, size=1000)

# 変換式に従って指数分布のサンプルを生成
ys = inv_integral_of_exp_dist(zs, 1.0)
```

```python
fig = plt.figure(1)
ax = fig.add_subplot(1, 1, 1)
ax.plot(xs, exps) # 指数分布の描画
ax.plot(zs, [0.2] * np.alen(zs), "o") # 一様分布のサンプル点
ax.plot(ys, [0.0] * np.alen(ys), "o") # 指数分布のサンプル点
 # 指数分布のサンプル点の正規化ヒストグラム
n, bins, patches = ax.hist(ys, bins=50, normed=True, range=(0.0, 10.0))
ax.legend(["exp distribution", "uniform samples", "exp samples", "exp samples hist"])
```

## コーシー分布(Cauchy distribution)の例 ##

$$ p(y) = \frac{1}{\pi}\frac{1}{1+y^2} $$

* 不定積分は計算すると$ (1/\pi) \tan^{-1}(y) + 1/2 $となりその逆関数は$y=\tan(\pi(z-1/2))$
    * 演習11.3

```python
def cauchy_dist(xs):
    return (1.0 / np.pi) * (1.0 / (1 + xs ** 2))

def inv_integral_of_cauchy_dist(zs):
    return np.tan(np.pi * (zs - 0.5))
```

```python
from numpy.random import RandomState

# コーシー分布
xs = np.linspace(-10.0, 10.0, 1000)
cauchys = cauchy_dist(xs)

# 0.0 ~ 1.0の一様分布に従うサンプルを1000点生成
# RandomStateの引数はseed
zs = RandomState(1).uniform(0.0, 1.0, size=1000)

# 変換式に従ってコーシー分布のサンプルを生成
ys = inv_integral_of_cauchy_dist(zs)
```

```python
fig = plt.figure(1)
ax = fig.add_subplot(1, 1, 1)
ax.plot(xs, cauchys) # コーシー分布の描画
ax.plot(zs, [0.2] * np.alen(zs), "o") # 一様分布のサンプル点
ax.plot(ys, [0.0] * np.alen(ys), "o") # コーシー分布のサンプル点
 # コーシー分布のサンプル点の正規化ヒストグラム
ax.hist(ys, bins=100, normed=True, range=(-10.0, 10.0))
ax.set_xlim([-10.0, 10.0])
ax.legend(["cauchy dist", "uni samples", "cauchy samples", "cauchy hist"], loc="best")
```

## 多変数への拡張 ##

* 変数変換のヤコビ行列式を以下のように用いる

$$ p(y_1,...,y_M) = p(z_1,...,z_M)\left| \frac{\partial(z_1,...,z_M)}{\partial(y_1,...,y_M)} \right| $$

## Box-Muller法(Box-Muller method) ##

* ガウス分布に従うサンプルを発生させる手法
* 正確にはMarsaglia法という

### アルゴリズム ###

* 1: 一様分布する乱数のペア$z_1, z_2 \in (-1,1)$を生成
    * (0,1)で一様分布する変数を2z-1と変換する
* 2: $z_1^2+z_2^2\leq 1$を満たさないペアを破棄
* 3: 単位円の内部で確率密度$p(z_1,z_2)=1/\pi$に従って一様分布する点の集合が得られる

```python
disp_figure("11.3")
```

```python
from numpy.random import RandomState

z1s = RandomState(1).uniform(-1.0, 1.0, size=1000)
z2s = RandomState(2).uniform(-1.0, 1.0, size=1000)

z_mask = ((z1s ** 2) + (z2s ** 2) <= 1.0)
z_pairs = np.array(zip(z1s, z2s))[z_mask]
print(np.alen(z_pairs))
```

```python
fig = plt.figure(1, figsize=(6, 6))
ax = fig.add_subplot(111)
ax.scatter(z_pairs[:,0], z_pairs[:,1])
ax.set_xlim([-1.0, 1.0])
ax.set_ylim([-1.0, 1.0])
```

### アルゴリズム続き ###

* 4: それぞれのペア$z_1,z_2$に対し以下を計算($r^2=z_1^2+z_2^2$)

$$ y_1 = z_1 \left( \frac{-2\ln r^2}{r^2}\right) ^{1/2} $$
$$ y_2 = z_2 \left( \frac{-2\ln r^2}{r^2}\right) ^{1/2} $$

* 5: $y_1$と$y_2$の同時分布が以下で与えられる
    * $y_1$と$y_2$は独立でありそれぞれが平均0、分散1のガウス分布に従う

$$ p(y_1, y_2) = p(z_1, z_2) \left| \frac{\partial (z_1, z_2)}{\partial (y_1, y_2)} \right| $$
$$ = \left[ \frac{1}{\sqrt{2\pi}\exp(-y_1^2/2)} \right] \left[ \frac{1}{\sqrt{2\pi}\exp(-y_2^2/2)} \right] $$

```python
def y_pairs_from_z_pairs(z_pairs):
    r_power = (z_pairs[:, 0] ** 2) + (z_pairs[:,1] ** 2)
    common_term = np.sqrt((-2.0 * np.log(r_power)) / r_power)
    return z_pairs * common_term[:, np.newaxis]

y_pairs = y_pairs_from_z_pairs(z_pairs)
```

```python
fig = plt.figure(1, figsize=(6, 6))
ax = fig.add_subplot(211)
_ = ax.hist(y_pairs[:,0], bins=50, range=(-5.0, 5.0))
ax = fig.add_subplot(212)
_ = ax.hist(y_pairs[:,1], bins=50, range=(-5.0, 5.0))
```

## 演習11.4 ##

上記の変数変換
$$ y_1 = z_1 \left( \frac{-2\ln r^2}{r^2}\right) ^{1/2} $$
$$ y_2 = z_2 \left( \frac{-2\ln r^2}{r^2}\right) ^{1/2} $$
を行うと、$(y_1, y_2)$が独立なガウス分布の積
$$ p(y_1, y_2) = \left[ \frac{1}{\sqrt{2\pi}\exp(-y_1^2/2)} \right] \left[ \frac{1}{\sqrt{2\pi}\exp(-y_2^2/2)} \right] $$
となることを示せ。

## 解答 ##

$$ z_1 = r \cos\theta $$
$$ z_1 = r \sin\theta $$

と変数変換を行うと

$$ y_1 = (-2\ln r^2)^{1/2} \cos\theta$$
$$ y_2 = (-2\ln r^2)^{1/2} \sin\theta$$

この時
$$ y_1^2 + y_2^2 = (-2\ln r^2)^{1/2} \iff \ln r^2 = - \frac{y_1^2 + y_2^2}{2}$$
$$ \iff r^2 = \exp\left(- \frac{y_1^2 + y_2^2}{2} \right) = \exp\left(- \frac{y_1^2}{2} \right)\exp\left(- \frac{y_2^2}{2} \right) $$

$$ p(y_1, y_2) = p(z_1, z_2) \left| \frac{\partial (z_1, z_2)}{\partial (y_1, y_2)} \right| = p(z_1, z_2) \left| \frac{\partial (z_1, z_2)}{\partial (r, \theta)} \right| \left| \frac{\partial (r, \theta)}{\partial (y_1, y_2)} \right| $$
であるため、それぞれのヤコビアンを計算することで$p(y_1, y_2)$を求める。

$$ \left| \frac{\partial (z_1, z_2)}{\partial (r, \theta)} \right| = \begin{vmatrix} \frac{\partial z_1}{\partial r} & \frac{\partial z_1}{\partial \theta} \\ \frac{\partial z_2}{\partial r} & \frac{\partial z_2}{\partial \theta} \end{vmatrix} = \begin{vmatrix} \cos\theta & -r\sin\theta \\ \sin\theta & r\cos\theta \end{vmatrix} = r $$
$$ \left| \frac{\partial (r, \theta)}{\partial (y_1, y_2)} \right| = \left| \frac{\partial (y_1, y_2)}{\partial (r, \theta)} \right|^{-1} $$
$$ =  \begin{vmatrix} -\frac{2}{r}\frac{1}{(-2\ln r^2)^{1/2}}\cos\theta & -(-2\ln r^2)^{1/2} \sin\theta \\ -\frac{2}{r}\frac{1}{(-2\ln r^2)^{1/2}}\sin\theta & (-2\ln r^2)^{1/2} \cos\theta \end{vmatrix}^{-1} $$
$$ = \left(- \frac{2}{r}\right)^{-1} = - \frac{r}{2} $$
従って
$$ p(y_1, y_2) = \frac{1}{\pi} \cdot r \cdot \frac{r}{2} = \frac{1}{2\pi}r^2 $$

以上より
$$ p(y_1, y_2) = \frac{1}{\pi} \cdot r \cdot \frac{r}{2} = \frac{1}{2\pi}r^2 = \frac{1}{\sqrt{2\pi}}\exp\left(- \frac{y_1^2}{2} \right)\frac{1}{\sqrt{2\pi}}\exp\left(- \frac{y_2^2}{2} \right) $$
よって示された。

* 1変数の分布について考え、その後に多次元への拡張について考える

## 問題設定 ##

* 標準的な分布ではない分布$p({\bf z})$からサンプリングしたいが直接は困難であると仮定
* 任意の与えられた${\bf z}$の値について$p({\bf z})$を求めることはある正規化定数を除いて容易
    * 実際によくあること
    * $\tilde{p}(z)$はすぐに求められるが$Z_p$は分からない
$$ p(z)=\frac{1}{Z_p}\tilde{p}(z) $$

## 棄却サンプリングの考え方 ##

* 容易にサンプルを抽出することができるより簡単な提案分布$q(z)$を利用
* 次に定数kに対し、zの全ての値に対して$kq(z) \geq \tilde{p}(z)$が成り立つようにその値を決める
    * $kq(z)$は比較関数と呼ばれる

## アルゴリズム ##

1. 乱数$z_0$を分布$q(z)$から生成
2. 乱数$u_0$を区間$[0, kq(z_0)]$上の一様分布から生成
    * $kq(z)$の曲線の下側で一様分布
3. $u_0 > \tilde{p}(z_0)$ならばサンプルを破棄、そうでなければ$u_0$が保持
    * 図11.4の灰色の領域では棄却される
4. 残った乱数のペアの集合は$\tilde{p}(z)$の曲線の下側で一様に分布し、対応するzの値は望んだ通り$p(z)$に従って分布

```python
disp_figure("11.4")
```

## サンプルの受理確率 ##

* zのもともとの値は分布$q(z)$から生成され、それらのサンプルは確率$\tilde{p}(z)/kq(z)$で受理
* 1つのサンプルが受理される確率は

$$ p(受理) = \int {\tilde{p}(z)/kq(z)}q(z)dz $$
$$ = \frac{1}{k} \int \tilde{p}(z) dz $$

* 棄却される点の割合は正規化されていない分布$\tilde{p}(z)$の下側の領域と曲線$kq(z)$の下側の領域との割合
    * それ故定数kはできるだけ小さくするべき

## ガンマ分布(gamma distribution)の例 ##

* ガンマ分布からのサンプリングを考える

$$ {\rm Gam}(z|a,b)=\frac{b^az^{a-1}\exp(-bz)}{\Gamma(a)} $$

* ガンマ分布はa>1の時図11.5に示すような釣鐘型の分布
    * 提案分布としては釣鐘型で前述の変換法を利用可能なコーシー分布が適している

```python
# 緑がガンマ分布、赤が提案分布であるコーシー分布
disp_figure("11.5")
```

## コーシー分布の一般化 ##

* どんな場合でもガンマ分布よりも小さい値を取らないように変換
    * 一様分布の確率変数yを$z=b\tan y + c$を用いて変換することで実現可能
    * これにより乱数は以下に従って分布
    
$$ q(z) = \frac{k}{1+(z-c)^2/b^2} $$

* 最小の棄却率は$c=a-1$と$b^2=2a-1$に設定し、$kq(z) \geq \tilde{p}(z)$の条件を満足させながら定数kをできるだけ小さくすることで得られる

## 演習 11.7 ##

yが区間[0, 1]上の一様分布に従うとしたとき、変数$z=b\tan y + c$が(11.16)で与えられるコーシー分布に従うことを示せ

### 解答 ###

$z=b\tan y + c$は想定する確率分布の不定積分の逆関数であるため、この逆関数を求めて微分すると求める確率分布となる。
従って
$$z=b\tan y + c \iff y = \arctan(\frac{z-c}{b})$$
を微分すると、
$$\frac{{\rm d}y}{{\rm d}z} = \frac{1}{b}\frac{1}{1+(z-c)^2/b^2} $$
$k=1/b$とすると(11.16)に一致する。

## 具体的なアルゴリズム ##

1. 一般化コーシー分布のパラメータを$c=a-1,b^2=2a-1$とおいて常にガンマ分布よりも大きくなるkを探す
2. [0,1]の一様分布を生成し一般化コーシー分布のサンプル点を$z=b\tan y + c$に基づき得る
3. 得られたサンプル点zに対応する一般化コーシー分布の確率値kq(z)を計算し、[0, kq(z)]の一様分布からサンプルuを生成、u <= p(z)であればzを受理

```python
from scipy.special import gamma

def gamma_dist(xs, a=1.0, b=1.0):
    return (b ** a) * (xs ** (a - 1)) * np.exp(- b * xs) / gamma(a)
    
def generalized_cauchy_dist(xs, c=0., b=1., k=1.):
    return k / (1 + ((xs - c) ** 2 / b ** 2))

# [0, 1]の一様分布を一般化されたコーシー分布に移す
# tanの中で[0, 1] -> [-pi/2, pi/2]という変換を挟んでいる
def uniform_to_cauchy(ys, c=0., b=1.):
    return b * np.tan(np.pi*(ys-0.5)) + c
```

```python
xs = np.linspace(0.0, 30.0, 1000)

# gamma分布、cauchy分布のパラメータ
ga, gb    = (7.5, 1.0)
cc, cb, ck = (ga - 1, np.sqrt(2.0 * ga - 1), 0.16)

fig = plt.figure(1)
ax = fig.add_subplot(111)
ax.plot(xs, gamma_dist(xs, ga, gb))
ax.plot(xs, generalized_cauchy_dist(xs, cc, cb, ck))
ax.plot(xs, generalized_cauchy_dist(xs, cc, cb, ck - 0.01))
ax.legend(["gamma", "generalized cauchy(k=0.16)", "generalized cauchy(k=0.15)"])
```

```python
# ck = 0.15〜0.16の範囲で常にガンマ分布よりも大きくなるようなkを探す
gs = gamma_dist(xs, ga, gb)
for k in np.arange(0.15, 0.16, 0.0001):
    if np.all(generalized_cauchy_dist(xs, cc, cb, k) > gs):
        print k
        break
```

```python
# gamma分布、cauchy分布のパラメータ
ga, gb    = (7.5, 1.0)
cc, cb, ck = (ga - 1, np.sqrt(2.0 * ga - 1), 0.1545)

fig = plt.figure(1)
ax = fig.add_subplot(111)
ax.plot(xs, gamma_dist(xs, ga, gb))
ax.plot(xs, generalized_cauchy_dist(xs, cc, cb, ck))
ax.legend(["gamma", "generalized cauchy"])
```

```python
from numpy.random import RandomState

# 0.0 ~ 1.0の一様分布に従うサンプルを5000点生成
ys = RandomState(1).uniform(0.0, 1.0, size=5000)

# 変換式に従ってコーシー分布のサンプルを生成
zs = uniform_to_cauchy(ys, cc, cb)
zs = zs[zs > 0.0] # 負のサンプルを除外
print(np.alen(zs))
```

```python
fig = plt.figure(1)
ax = fig.add_subplot(111)
ax.plot(xs, gamma_dist(xs, ga, gb))
ax.plot(xs, generalized_cauchy_dist(xs, cc, cb, ck))
# 一般化コーシー分布のヒストグラム、
# ここでは横軸の値のみが重要であり、確率値そのものは数式から計算するため、
# ここではヒストグラムの概形がコーシー分布に近いかだけに着目すれば良い
ax.hist(zs, bins=100, normed=True, range=(0., 30.0))
ax.set_xlim([0.0, 30.0])
```

```python
# コーシー分布の各サンプルに対して[0.0, kq(z)]の一様分布を計算し、棄却判定を行う
zs_accepted = []
for z in zs:
    uni_upper = generalized_cauchy_dist(z, cc, cb, ck)
    # ここでseedを固定すると精度がものすごい悪くなる
    u = RandomState().uniform(0.0, uni_upper, size=1)[0]
    g = gamma_dist(z, ga, gb)
    if u <= g:
        zs_accepted.append(z)

print(float(np.alen(zs_accepted)) / float(np.alen(zs)))
```

```python
fig = plt.figure(1)
ax = fig.add_subplot(111)
ax.plot(xs, gamma_dist(xs, ga, gb))
ax.plot(xs, generalized_cauchy_dist(xs, cc, cb, ck))
ax.hist(zs_accepted, bins=100, normed=True, range=(0., 30.0))
ax.set_xlim([0.0, 30.0])
```

## 適応的棄却サンプリング ##

* 棄却サンプリングにおいて適切な包絡分布$q(z)$を解析的な形で決定することは難しい
* 別のアプローチとして$p(z)$の観測値にもとづいてその場で包絡関数を構築
    * $p(z)$が対数凹($\ln p(z)$の導関数が非増加関数)である場合に包絡関数の構築は容易
* 図11.6に構築方法の図解有

```python
disp_figure("11.6")
```

## 構築方法 ##

* 関数$\ln p(z)$とその勾配がグリッド点の何らかの初期集合で求められる
* 結果として得られる接線の交点が包絡関数を構築するために用いられる
* 包絡分布の対数は線形関数の連なりであるため、包絡関数自体は区分的な指数分布となる
    * $\tilde{z}_{i-1,i}$は$z_{i-1}, z_i$における接線の交点、$\lambda_i$は$z_i$での接線の傾き、$k_i$は対応するオフセット
    * サンプリングは容易(演習11.9)

$$ q(z) = k_i\lambda_i\exp\{-\lambda_i(z-z_i)\} $$
$$ \tilde{z}_{i-1,i} < z \leq \tilde{z}_{i,i+1} $$


## 演習11.9 ##

どうやればサンプリングが可能であるか考えよ

### 解答 ###

各区間ごとに異なる指数分布に従う訳なので、各区間ごとに異なるパラメータの変換関数を適用すれば良い。
変換関数は指数分布を不定積分し、その逆関数を求めることで得られる。

## アルゴリズム ##

1. 区分的な指数分布からサンプルを抽出
2. 通常の棄却基準が適用され、受理されれば所望の分布からの抽出となり、棄却されればグリッド点の集合に追加、新しい接線が計算され包絡関数が改良
3. グリッド点の個数が増えるにつれ包絡関数が求めたい分布をより良く近似

## 適応的棄却Metropolisサンプリング ##

* adaptive rejection Metropolis sampling
* アルゴリズムの変形版
    * 各棄却サンプリングステップの後に11.2.2節のMetropolis-Hastingsステップを置く
    * 対数凹でない分布に対しての拡張

## 棄却サンプリングの実用上の価値 ##

* 持たせるには棄却率が最小になるように比較関数を想定する分布に近づける必要

## 高次元空間における棄却サンプリング ##

* 平均0, 共分散$\sigma_p^2{\bf I}$の多変量ガウス分布からのサンプリングを考える
* 提案分布を平均0, 共分散$\sigma_q^2{\bf I}$とおく
* $kq(z) \geq p(z)$となるためには$\sigma_q^2 \geq \sigma_p^2$が必要

### D次元空間での受理率 ###

* D=1の場合は図11.7
* kの最適値は$(\sigma_q/\sigma_p)^D$
* 受理率はp(z)以下の体積とkq(z)以下の体積の比であり、ここでは両方の分布が正規化されているため1/k
* 1/k=$(\sigma_p/\sigma_q)^D$なので、D=1000次元で提案分布の分散が1%しか大きくない場合でも$\left(\sigma_p/(1.01*\sigma_q)\right)^{1000} \simeq 1/20000$
    * 次元数に対して指数的に減少

### より現実的な例の場合 ###

* 先の例は比較関数は求めたい分布に近い分布であるため現実的ではない
* 実際に求めたい分布は多峰性で鋭いピークを持つことが多い
    * この求めたい分布に対応する良い提案分布を求めることは困難

## 棄却サンプリングの特徴 ##

* 次元数に対して受理率が指数的に減少
    * 高次元には適していない
* 高次元空間におけるサンプリングのためのより洗練されたアルゴリズムにおいて、サブルーチンとしての役割は持てる
